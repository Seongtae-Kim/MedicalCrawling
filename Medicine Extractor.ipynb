# Medicine Extractor

```python
from selenium.webdriver.firefox.firefox_binary import FirefoxBinary
from selenium.webdriver.common.keys import Keys
from selenium import webdriver
from collections import defaultdict
from lxml import etree as ET
from lxml.builder import E
```

```python
url = "http://www.amc.seoul.kr/asan/healthinfo/druginfo/drugInfoList.do?pageIndex="
# 1 - 277

urlList = []
for i in range(277):
    urlList.append(url + str(i+1))
```

```python
class medicine:
    description=defaultdict()
```

```python
def medicine_extractor(urlList):
    medicineList = []
    for url in urlList:
        browser.get(url)
        tabList = []
        lf = browser.find_elements_by_class_name("listCont")
        
        if len(lf) != 0:
            tab_contents = lf[0].find_elements_by_tag_name("a")
            for tab_content in tab_contents:
                tabList.append(tab_content.get_attribute("href"))
            content_extractor(tabList, medicineList)
    return medicineList

def content_extractor(tabList, medicineList):
    
    for tab in tabList:
        browser.get(tab)
        drugName = browser.find_elements_by_class_name("drugName")
        if len(drugName) != 0:
            m = medicine()
            attacher = {}
            attacher["drug_name"] = drugName[0].text
            i=0
            for t in browser.find_elements_by_tag_name("td"):
                i += 1
                if i == 1: # 성분명
                    attacher["component"] = t.text
                if i == 2: # 처방명(한글)
                    attacher["prescription_ko"] = t.text
                if i == 3: # 처방명(영문)
                    attacher["prescription_en"] = t.text
                if i == 4: # 제조사명
                    attacher["manufacturer"] = t.text
                if i == 5: # 함량
                    attacher["content"] = t.text
                if i == 6: # 모양
                    attacher["shape"] = t.text
                if i == 7: # 분할선
                    attacher["cleavage_line"] = t.text
                if i == 8: # 색상1
                    attacher["color1"] = t.text
                if i == 9: # 색상2
                    attacher["color2"] = t.text
                if i ==10: # 표면글자(앞)
                    attacher["surface_letter_front"] = t.text
                if i ==11: # 표면글자(뒤)
                    attacher["surface_letter_back"] = t.text
                if i ==12: # 단위
                    attacher["unit"] = t.text
                if i ==13: # 제형분류
                    attacher["route_of_administration"] = t.text
                if i ==14: # 효능
                    attacher["effect"] = t.text
                if i ==15: # 부작용
                    attacher["side_effect"] = t.text
                if i ==16: # 의약품상호작용
                    attacher["interaction"] = t.text
                if i ==17: # 주의사항
                    attacher["caution"] = t.text
                if i ==18: # 보관방법
                    attacher["storage_solution"] = t.text
            
            m.description = attacher
            medicineList.append(m)    
            print(len(medicineList),"개 완료: ", medicineList[-1].description["drug_name"])
        
```

```python
browser = webdriver.Firefox()
mlist = medicine_extractor(urlList)
```

```python
whole = ""
for medicine in mlist:
    txt= ""
    txt = txt + "\n"
    txt = txt + "의약품명: " + medicine.description["drug_name"] + "\n"
    txt = txt + "성분명: " + medicine.description["component"] + "\n"
    txt = txt + "처방명(한글): " + medicine.description["prescription_ko"] + "\n"
    txt = txt + "처방명(영문): " + medicine.description["prescription_en"] + "\n"
    txt = txt + "제조사명: " + medicine.description["manufacturer"] + "\n"
    txt = txt + "함량: " + medicine.description["content"] + "\n"
    txt = txt + "모양: " + medicine.description["shape"] + "\n"
    txt = txt + "분할선: " + medicine.description["cleavage_line"] + "\n"
    txt = txt + "색상1: " + medicine.description["color1"] + "\n"
    txt = txt + "색상2: " + medicine.description["color2"] + "\n"
    txt = txt + "표면글자(앞): " + medicine.description["surface_letter_front"] + "\n"
    txt = txt + "표면글자(뒤): " + medicine.description["surface_letter_back"] + "\n"
    txt = txt + "단위: " + medicine.description["unit"] + "\n"
    txt = txt + "제형분류: " + medicine.description["route_of_administration"] + "\n"
    if "effect" in medicine.description.keys():
        txt = txt + "효능: " + medicine.description["effect"] + "\n"
    if "side_effect" in medicine.description.keys():
        txt = txt + "부작용: " + medicine.description["side_effect"] + "\n"
    if "interaction" in medicine.description.keys():
        txt = txt + "의약품상호작용: " + medicine.description["interaction"] + "\n"
    if "caution" in medicine.description.keys():
        txt = txt + "주의사항: " + medicine.description["caution"] + "\n"
    if "storage_solution" in medicine.description.keys():
        txt = txt + "보관방법: " + medicine.description["storage_solution"] + "\n"
    txt = txt + "\n"
    whole = whole + txt 
    with open("./output/medicine_ "+ medicine.description["drug_name"].split("\n")[0].replace("/", "_").replace(" ", "_") +".txt", "w") as text_file:
        text_file.write(txt)

with open("./whole/medicines.txt", "w") as text_file:
    text_file.write(whole)
```

```python
medicines_xml = E("medicines")
for medicine in mlist:
    drug_name = E("drug_name", medicine.description["drug_name"])
    component = E("component", medicine.description["component"])
    prescription_ko = E("prescription_ko", medicine.description["prescription_ko"])
    prescription_en = E("prescription_en", medicine.description["prescription_en"])
    manufacturer = E("manufacturer", medicine.description["manufacturer"])
    content = E("content", medicine.description["content"])
    shape = E("shape", medicine.description["shape"])
    cleavage_line = E("cleavage_line", medicine.description["cleavage_line"])
    color1 = E("color1", medicine.description["color1"])
    color2 = E("color2", medicine.description["color2"])
    surface_letter_front = E("surface_letter_front", medicine.description["surface_letter_front"])
    surface_letter_back = E("surface_letter_back", medicine.description["surface_letter_back"])
    unit = E("unit", medicine.description["unit"])
    route_of_administration = E("route_of_administration", medicine.description["route_of_administration"])   
    if "effect" in medicine.description.keys():
        effect = E("effect",  medicine.description["effect"])
    if "side_effect" in medicine.description.keys():
        side_effect = E("side_effect", medicine.description["side_effect"])
    if "interaction" in medicine.description.keys():
        interaction = E("interaction", medicine.description["interaction"])
    if "caution" in medicine.description.keys():
        caution = E("caution", medicine.description["caution"])
    if "storage_solution" in medicine.description.keys():
        storage_solution = E("storage_solution", medicine.description["storage_solution"])
    
        
    medicine_xml = E("medicine", drug_name, component, prescription_ko, prescription_en,
                     manufacturer, content, shape, cleavage_line, color1, color2,
                     surface_letter_front, surface_letter_back, unit, route_of_administration,
                     effect, side_effect, interaction, caution, storage_solution)
    medicines_xml.append(medicine_xml)

declaration = """<?xml version="1.0" encoding="utf-8" standalone="yes"?>
<!DOCTYPE medicines [
<!ELEMENT medicines (medicine)*>
<!ELEMENT medicine (drug_name, component, prescription_ko, prescription_en, manufacturer, content, shape, cleavage_line, color1, color2, surface_letter_front, surface_letter_back, unit, route_of_administration, effect, side_effect, interaction, caution, storage_solution)*>    
<!ELEMENT drug_name (#PCDATA)>
<!ELEMENT component (#PCDATA)>
<!ELEMENT prescription_ko (#PCDATA)>
<!ELEMENT prescription_en (#PCDATA)>
<!ELEMENT manufacturer (#PCDATA)>
<!ELEMENT content (#PCDATA)>
<!ELEMENT shape (#PCDATA)>
<!ELEMENT cleavage_line (#PCDATA)>
<!ELEMENT color1 (#PCDATA)>
<!ELEMENT color2 (#PCDATA)>
<!ELEMENT surface_letter_front (#PCDATA)>
<!ELEMENT surface_letter_back (#PCDATA)>
<!ELEMENT unit (#PCDATA)>
<!ELEMENT route_of_administration (#PCDATA)>
<!ELEMENT effect (#PCDATA)>
<!ELEMENT side_effect (#PCDATA)>
<!ELEMENT interaction (#PCDATA)>
<!ELEMENT caution (#PCDATA)>
<!ELEMENT storage_solution (#PCDATA)>
]>
<!--Seongtae Kim 2020-06-06 https://github.com/Seongtae-Kim -->
"""

dec_byte = str.encode(declaration)
medicines_tree = ET.ElementTree(medicines_xml)
end_com= str.encode("<!--Seongtae Kim 2020-06-06 https://github.com/Seongtae-Kim -->")

with open("./xml/medicines.xml", "wb") as f:
    f.write(dec_byte)
    f.write(ET.tostring(medicines_tree, xml_declaration=False, encoding="utf-8", pretty_print=True))
    f.write(end_com)
```
